### Synthesis route prediction: Some New features
- Carbonate, Hyrdoxide, Nitrate precursors.
- Temperature (and pressure) control via gas phase
- ICSD materials
- Literature extracted precursors.
- Custom phase additions

```python
from rxn.route import SynthesisRoutes
mp_id = 'mp-5229'
```

Let's look at SrTiO3 solid-state synthesis, explicitly include rutile-TiO2, include carbonates, and ammbient pressure.

```python
router = SynthesisRoutes(mp_id, add_element='C', explicit_includes=['mp-390'], pressure=None)
```

Possible reactions:

```python
len(router.reactions)
```

Let's get a recommendation plot -- barrier vs # competing phases. We also turn on "gas release". Carbonates mean we will have CO/CO2 release.

```python
router.recommend_routes(temperature=1500, pressure=None, allow_gas_release=True)
```

```python
router.pressure
```

Too many reactions above. Many are Sr-Ti-O phases. We can simplify:

```python
router.recommend_routes(temperature=1500, pressure=None, allow_gas_release=True, max_component_precursors=2)
```

We can further filter down to only "literature extracted precursors"

```python
router.recommend_routes(temperature=1500, allow_gas_release=True, show_known_precursors_only=True)
```
